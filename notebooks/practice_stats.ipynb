### 中心極限定理

```python
# グラフ描写の初期設定
fig = plt.figure(figsize=(10, 15))
axes = fig.subplots(2,1)

def hist_sample_means(ax, samples):
    sample_means = np.mean(samples, axis=1)
    ax.hist(sample_means, bins=100, density=True)


def plot_actual_normal(ax, mena, variance, n):
    # stats.norm()にわたすのは平均と標準偏差
    normal = stats.norm(mean, np.sqrt(variance/n))
    # 下側確率0.999の点
    x_max = normal.ppf(0.999)
    # 下側確率0.001の点
    x_min = normal.ppf(0.001)
    x = np.linspace(x_max, x_min, 10000)
    ax.plot(x, normal.pdf(x))

# ポアソン分布 E(X)=λ、V(X)=λ
n = 100000
lambda_ = 3
mean = lambda_
siguma_squared = lambda_
poisson_samples = stats.poisson(lambda_).rvs(size=(500, n))

hist_sample_means(axes[0], poisson_samples)
plot_actual_normal(axes[0], mean, siguma_squared, n)
axes[0].set_title('Poisson Distribution')


# カイ二乗分布 E(X)=K、V(X)=2k
n = 100000
k = 5
mean = k
siguma_squared = 2*k
poisson_samples = stats.chi2(k).rvs(size=(500, n))
hist_sample_means(axes[1], poisson_samples)
plot_actual_normal(axes[1], mean, siguma_squared, n)
axes[1].set_title('Chi-squared Distribution')

plt.show()
```
